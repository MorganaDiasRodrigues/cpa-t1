```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from PIL import Image
from selenium.webdriver.support.ui import WebDriverWait
import requests
import shutil
import os
from os import path
import json
```

```python
path_to_folder = './img_posters'

if not path.exists(path_to_folder):
    os.mkdir(path_to_folder)


def download_poster(poster_url, title):
    response = requests.get(poster_url, stream=True)
    with open(f'{path_to_folder}/{title}.png', 'wb') as f:
        shutil.copyfileobj(response.raw, f)

    return f'{path_to_folder}/{title}.png'
```

```python
path_to_folder = './img_posters'

if not path.exists(path_to_folder):
    os.mkdir(path_to_folder)


def download_poster(poster_url, title):
    response = requests.get(poster_url, stream=True)
    with open(f'{path_to_folder}/{title}.png', 'wb') as f:
        shutil.copyfileobj(response.raw, f)

    return f'{path_to_folder}/{title}.png'


PATH = 'chromedriver.exe'
driver = webdriver.Chrome(PATH)
driver.maximize_window()

driver.get(url="https://www.imdb.com/chart/top/?ref_=nv_mv_250")
WebDriverWait(driver, 10)

numero_filme = 1
filmes = []


while numero_filme <= 250:
    filme = {}
    print("Filme: ", numero_filme)
    titulo = driver.find_element_by_xpath(f'//*[@id="main"]/div/span/div/div/div[3]/table/tbody/tr[{numero_filme}]/td[2]/a').text
    filme.update({"titulo":titulo})
    print(titulo)
    ano = driver.find_element_by_xpath(f'//*[@id="main"]/div/span/div/div/div[3]/table/tbody/tr[{numero_filme}]/td[2]/span')
    filme.update({"ano":ano.text})
    print(ano.text)
    url_poster = driver.find_element_by_xpath(f'//*[@id="main"]/div/span/div/div/div[3]/table/tbody/tr[{numero_filme}]/td[1]/a/img')
    filme.update({"url_poster":url_poster.get_attribute("src")})
    print(url_poster.get_attribute("src"))
    img_poster = download_poster(url_poster.get_attribute("src"), titulo)
    filme.update({"img_poster":img_poster})
    nota = driver.find_element_by_xpath(f'//*[@id="main"]/div/span/div/div/div[3]/table/tbody/tr[{numero_filme}]/td[3]/strong') 
    filme.update({"nota":nota.text})
    print(nota.text)
    numero_filme+=1
    filmes.append(filme)

PATH = 'chromedriver.exe'
driver = webdriver.Chrome(PATH)
driver.maximize_window()

numero_filme = 1

while numero_filme <= 250:
    filme = {}
    driver.get(url="https://www.imdb.com/chart/top/?ref_=nv_mv_250")
    # Clicando no filme
    driver.find_element_by_xpath(f'//*[@id="main"]/div/span/div/div/div[3]/table/tbody/tr[{numero_filme}]/td[2]/a').click()
    print("Filme: ", numero_filme)
    gender_list = driver.find_element_by_xpath('//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/div[1]/div')
    filme.update({"gender_list":gender_list.text})
    print(gender_list.text)
    popularity = driver.find_element_by_xpath('//*[@data-testid="hero-rating-bar__popularity__score"]')
    filme.update({"popularity":popularity.text})
    print("POPULARIDADE: ", popularity.text)
    diretor = driver.find_element_by_xpath('//*[@class="ipc-metadata-list-item__content-container"]/ul/li/a').text
    filme.update({"diretor":diretor})
    print(diretor)
    numero_filme +=1
    filmes.append(filme)

with open('filmes.json', 'w') as outfile:
    json.dump(filmes, outfile)


```
